```python
import io
import json
import numpy as np
import pandas as pd
from pathlib import Path

# Load data from CSV files
datasets_paths = {
    "UAE":Path(r"..\data\where\UAE.csv"),
    "US":Path(r"..\data\where\USA.csv"),
    "brain":Path(r"..\data\brain_tumor_dataset.csv"),
    "brain2":Path(r"..\data\brain_tumor_dataset2.csv"),
    "breast":Path(r"..\data\breast_cancer_dataset.csv"),
    "lung":Path(r"..\data\lung_cancer_dataset.csv"),
    "leukemia":Path(r"..\data\leukemia_dataset.csv"),
    "oral":Path(r"..\data\oral_cancer_dataset.csv"),
    "prostate":Path(r"..\data\prostate_cancer_dataset.csv"),
    "skin":Path(r"..\data\skin_cancer_dataset.csv"),
    "thyroid":Path(r"..\data\thyroid_cancer_dataset.csv")
}
datasets_paths
```

```python
for datas in datasets_paths.values():
    buffer = io.StringIO()
    df = pd.read_csv(datas)
    print(f"Dataset: {df.describe().to_string()}")
    with open(Path(*datas.parts[:-1]) / f"{datas.name.split(".")[0]}.description", "w") as f :
        f.write(df.describe().to_string())
    with open(Path(*datas.parts[:-1]) / f"{datas.name.split(".")[0]}.infos", "w") as f :
        df.info(buf=buffer)
        s = buffer.getvalue()
        f.write(s)
```
