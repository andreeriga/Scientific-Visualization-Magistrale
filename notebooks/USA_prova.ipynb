```python
def plotUSA(usa_merged, col, figsize=(12,8), cmap="viridis"):

    fig, ax = plt.subplots(figsize=figsize)

    try:
        usa_merged.plot(column=col, legend=True, ax=ax, cmap=cmap, edgecolor="black", linewidth=0.2)
    except Exception:
        usa_merged.plot(ax=ax)

    ax.set_title(f"USA - {col}")
    ax.axis("off")

    return fig, ax
```
